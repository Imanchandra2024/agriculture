**State which is more prone to agriculture**

```python
state_counts = df['state'].value_counts()

# create a bar plot
plt.figure(figsize=(12,6))
sns.barplot(x=state_counts.index,y=state_counts.values)
plt.xticks(rotation = 90)
plt.xlabel('state')
plt.ylabel('counts')
plt.title('state vs count')
plt.tight_layout()
plt.show()
```

**top  state which are involved in agriculture**

```python
plt.pie(df['state'].value_counts().to_list()[:10],labels=df['state'].value_counts().index[:10],radius=1.8,autopct="%0.2f%%")

plt.show()
```

```python
df.head()
```

district in perticular state which are more prone towards agriculture


```python
# Get the top 10 states
top_10_states = df['state'].value_counts().head(10).index

# Iterate through the top 10 states and create a bar plot for each
for state in top_10_states:
    district_counts = df[df['state'] == state]['district'].value_counts()

    # create a bar plot
    plt.figure(figsize=(12, 6))
    sns.barplot(x=district_counts.index, y=district_counts.values)
    plt.xticks(rotation=90)
    plt.xlabel('District')
    plt.ylabel('Count')
    plt.title(f'District Counts in {state}')
    plt.tight_layout()
    plt.show()
```

**commodity which are produced in large aomunt**

```python
values = df['commodity'].value_counts().to_list()[:10]
labels = df['commodity'].value_counts().index[:10]

plt.figure(figsize=(15,5))
sns.barplot(x=values, y=labels)
plt.show()
```

```python
numerical_df = df.select_dtypes(include=np.number)
sns.heatmap(numerical_df.corr(),annot=True)

plt.show()
```

```python
# Create a pair plot of the numerical price columns
sns.pairplot(df[['min_price', 'max_price', 'modal_price']])
plt.suptitle('Pair Plot of Price Columns', y=1.02)
plt.show()
```

```python
# Group by state and market and count commodities
commodity_counts = df.groupby(['state','market'])['commodity'].count().sort_values(ascending=False)

# Display the top 20 combinations
top_20_commodity_counts = commodity_counts.head(20)

# Create labels for the x-axis from the MultiIndex
labels = [f'{state}, {market}' for state, market in top_20_commodity_counts.index]

# Create a bar plot for the top 20 combinations
plt.figure(figsize=(14, 7))
sns.barplot(x=labels, y=top_20_commodity_counts.values)
plt.xticks(rotation=90)
plt.xlabel('State, Market')
plt.ylabel('Number of Commodities')
plt.title('Top 20 Commodity Counts per State and Market')
plt.tight_layout()
plt.show()
```
